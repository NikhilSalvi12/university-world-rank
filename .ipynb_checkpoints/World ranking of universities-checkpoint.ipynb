Importing the libraries for spliting data and making models

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X = data.drop(columns="world_rank")

y = data.world_rank.copy()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.03,random_state=42)

print(f"Shape of X train : {X_train.shape},\n Shape of X test :{X_test.shape} \
\n Shape of y train : {y_train.shape},\n Shape of y test :{y_test.shape}")
```

```python
norm = StandardScaler(with_std=True)

X_normtrain = norm.fit_transform(X_train)
X_normtest = norm.transform(X_test)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()

linear_model.fit(X_normtrain,y_train)
```

```python
pred_y_linear = linear_model.predict(X_normtest)
```

```python
linear_RMSE = mean_squared_error(y_test,pred_y_linear,squared=False)

print("RMSE: ",linear_RMSE)
```

# Poly features

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
score_df = pd.DataFrame(columns=["degree","RMSE","train error"])
for i in range(2,7):
    poly = PolynomialFeatures(degree=i)
    X_polytrain = poly.fit_transform(X_normtrain)
    X_polytest = poly.transform(X_normtest)
    
    model = LinearRegression()
    model.fit(X_polytrain,y_train)
    predicted = model.predict(X_polytest)
    RMSE = mean_squared_error(y_test,predicted,squared=False)
    score_df = score_df.append({"degree":i,"RMSE":RMSE,"train error":mean_squared_error(y_train,model.predict(X_polytrain),squared=False)},ignore_index=True)
    print("+",end="")
print("Loop Complete")
print(score_df)
```

**3 is the optimal degree that gives us lowest RMS error with lower over fitting with linear regression model**

Lets try with the Decision Tree model

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dectree_model = DecisionTreeRegressor(random_state=42)

dectree_model.fit(X_normtrain,y_train)
```

```python
tree_pred = dectree_model.predict(X_normtest)

tree_RMSE = mean_squared_error(y_test,tree_pred,squared=False)

print(f"RMSE for Decision Tree: {tree_RMSE}")
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(random_state=42)

rf_model.fit(X_normtrain,y_train)
```

```python
rf_pred = rf_model.predict(X_normtest)

rf_RMSE = mean_squared_error(y_test,rf_pred,squared=False)

print(f"RMSE for Random Forest: {rf_RMSE},\n training RMSE: {mean_squared_error(y_train,rf_model.predict(X_normtrain),squared=False)}")
```

**Let's try some boosting algorithms**

# Adaptive Boosting

```python
from sklearn.ensemble import AdaBoostRegressor

ad_score = pd.DataFrame(columns=["estimators","RMSE","train error"])
for i in range(50,2000,50):
    ad_model = AdaBoostRegressor(random_state=42,n_estimators=i)

    ad_model.fit(X_normtrain,y_train)
    
    ad_pred = ad_model.predict(X_normtest)

    ad_RMSE = mean_squared_error(y_test,ad_pred,squared=False)
    
    ad_score = ad_score.append({"estimators":i,"RMSE":ad_RMSE,"train error":mean_squared_error(y_train,ad_model.predict(X_normtrain),squared=False)},ignore_index=True)
    print("+",end="")
print("Loop complete")
```

```python
plt.figure(figsize=[16,8])
plt.plot(ad_score.estimators,ad_score.RMSE,"*-",label="testing")
plt.plot(ad_score.estimators,ad_score["train error"],"o-",label="training")
plt.title("Adaptive boosting")
plt.legend()
plt.grid()
plt.show()
```

Adaptive boosting doesn't give us the required results till now the random forest regression model has given us the best result lets try with another boosting algoritm

# Gradient Boost

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_score = pd.DataFrame(columns=["estimators","RMSE","train error"])
for i in range(100,1000,50):
    gb_model = GradientBoostingRegressor(criterion='mse',random_state=42,n_estimators=i)

    gb_model.fit(X_normtrain,y_train)
    
    gb_pred = gb_model.predict(X_normtest)

    gb_RMSE = mean_squared_error(y_test,gb_pred,squared=False)
    
    gb_score = gb_score.append({"estimators":i,"RMSE":ad_RMSE,"train error":mean_squared_error(y_train,gb_model.predict(X_normtrain),squared=False)},ignore_index=True)
    print("+",end="")
print("Loop complete")
```

```python
plt.figure(figsize=[16,8])
plt.plot(gb_score.estimators,gb_score.RMSE,"*-",label="testing")
plt.plot(gb_score.estimators,gb_score["train error"],"o-",label="training")
plt.title("Grandient boosting")
plt.legend()
plt.grid()
plt.show()
```

**Let's finalize the model that gives us the least error that is Forest Regressor**
